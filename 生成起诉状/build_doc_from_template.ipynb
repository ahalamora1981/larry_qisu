```python
df = pd.read_excel('info.xlsx', dtype=str)
```

```python
df = df.dropna(subset=["性别"]).reset_index().drop(columns=["index"])
```

```python
COL_NAMES = {
    "A": "公司名称",
    "B": "手别",
    "C": "案件id",
    "D": "用户ID",
    "E": "用户名",
    "F": "用户姓名",
    "G": "身份证号码",
    "H": "性别",
    "I": "民族",
    "J": "身份证地址",
    "K": "注册手机号",
    "L": "列表ID",
    "M": "合同号",
    "N": "资方机构",
    "O": "融担公司",
    "P": "合同金额",
    "Q": "放款日期",
    "R": "最后一期应还款日",
    "S": "借款期数",
    "T": "利率",
    "U": "逾期开始日期",
    "V": "上一个还款日期",
    "W": "列表逾期天数",
    "X": "待还金额",
    "Y": "待还本金",
    "Z": "待还费用",
    "AA": "数据提取日",
    "AB": "合并 代偿回购总额",
    "AC": "合并 代偿回购本金",
    "AD": "合并 代偿回购利息",
    "AE": "合并 代偿回购罚息",
    "AF": "是否可诉",
    "AG": "最晚代偿回购时间",
    "AH": "委派时间",
}
```

```python
from docx import Document
from docx.shared import Pt
from docx.oxml.ns import qn
from datetime import datetime, timedelta

from tqdm import tqdm
```

```python
for i in tqdm(range(df.shape[0])):
    doc = Document('template.docx')
    
    person = df['用户姓名'][i]
    contract_id = df['合同号'][i]
    
    # 判断担保公司
    company_fullname = df[df['合同号']==contract_id]['融担公司'].tolist()[0]
    
    if '福建智云' in company_fullname:
        company = '福建智云'
    elif '海南申信' in company_fullname:
        company = '海南申信'
    else:
        company = '上海耳序'
            
    file_name = f'{person}起诉状{contract_id[-4:]}-拓棱特-诉状-{company}-电子.docx'
    
    for j in range(len(doc.paragraphs)):
        text = doc.paragraphs[j].text
        
        if "【F】" in text:
            text = text.replace("【F】", df[COL_NAMES["F"]][i])
            
        if "【H】" in text:
            text = text.replace("【H】", df[COL_NAMES["H"]][i])
            
        if "【I】" in text:
            text = text.replace("【I】", df[COL_NAMES["I"]][i] + "族")

        if "【通过G读取】" in text:
            id_card = df[COL_NAMES["G"]][i]
            year = id_card[6:10]
            month = id_card[10:12]
            day = id_card[12:14]
            text = text.replace("【通过G读取】", f"{year}年{month}月{day}日")
            
        if "【J】" in text:
            text = text.replace("【J】", df[COL_NAMES["J"]][i])
            
        if "【K】" in text:
            text = text.replace("【K】", df[COL_NAMES["K"]][i])
            
        if "【G】" in text:
            text = text.replace("【G】", df[COL_NAMES["G"]][i])
            
        if "【Y】" in text:
            data_Y = float(df[COL_NAMES["Y"]][i])
            text = text.replace("【Y】", f"{data_Y:.2f}")
            
        if "【AD】" in text:
            data_AD = float(df[COL_NAMES["AD"]][i])
            text = text.replace("【AD】", f"{data_AD:.2f}")
            
        if "【AG】" in text:
            data_AG = df[COL_NAMES["AG"]][i].split(" ")[0]
            data_AG_date = datetime.strptime(data_AG, "%Y-%m-%d")
            year = data_AG_date.year
            month = data_AG_date.month
            day = data_AG_date.day
            text = text.replace("【AG】", f"{year}年{month}月{day}日")
            
        if "【AG+1】" in text:
            data_AG = df[COL_NAMES["AG"]][i].split(" ")[0]
            data_AG_date = datetime.strptime(data_AG, "%Y-%m-%d")
            data_AG_date = data_AG_date + timedelta(days=1)
            year = data_AG_date.year
            month = data_AG_date.month
            day = data_AG_date.day
            text = text.replace("【AG+1】", f"{year}年{month}月{day}日")
            
        if "【写状书当天】" in text:
            today = datetime.today()
            year = today.year
            month = today.month
            day = today.day
            text = text.replace("【写状书当天】", f"{year}年{month}月{day}日")
        
        if """【=Y*24%/360*DAYS("2024-4-3",AG)】""" in text:
            data_Y = float(df[COL_NAMES["Y"]][i])
            rate = 0.24
            data_AG = df[COL_NAMES["AG"]][i].split(" ")[0]
            data_AG_date = datetime.strptime(data_AG, "%Y-%m-%d")
            days = (datetime.today() - data_AG_date).days
            
            amount = ((data_Y * rate) / 360) * days
            text = text.replace("""【=Y*24%/360*DAYS("2024-4-3",AG)】""", f"{amount:.2f}")
            
        if """【Y+AD+Y*24%/360*DAYS("2024-4-3",AG)】""" in text:
            data_Y = float(df[COL_NAMES["Y"]][i])
            data_AD = float(df[COL_NAMES["AD"]][i])
            rate = 0.24
            data_AG = df[COL_NAMES["AG"]][i].split(" ")[0]
            data_AG_date = datetime.strptime(data_AG, "%Y-%m-%d")
            days = (datetime.today() - data_AG_date).days
            
            
            amount = data_Y + data_AD + (((data_Y * rate) / 360) * days) 
            text = text.replace("""【Y+AD+Y*24%/360*DAYS("2024-4-3",AG)】""", f"{amount:.2f}")
            
        if "【Q】" in text:
            data_Q = df[COL_NAMES["Q"]][i].split(" ")[0]
            data_Q_date = datetime.strptime(data_Q, "%Y-%m-%d")
            year = data_Q_date.year
            month = data_Q_date.month
            day = data_Q_date.day
            text = text.replace("【Q】", f"{year}年{month}月{day}日")
            
        if "【N（excel表里是简称须找到对应的全称）】" in text:
            text = text.replace("【N（excel表里是简称须找到对应的全称）】", df[COL_NAMES["N"]][i])

        if "【M】" in text:
            text = text.replace("【M】", df[COL_NAMES["M"]][i])
        
        if "【O】" in text:
            text = text.replace("【O】", df[COL_NAMES["O"]][i])
            
        if "【P】" in text:
            data_P = float(df[COL_NAMES["P"]][i])
            text = text.replace("【P】", f"{data_P:.2f}")
            
        if "【S】" in text:
            text = text.replace("【S】", df[COL_NAMES["S"]][i])
            
        if "【T】" in text:
            data_P = float(df[COL_NAMES["T"]][i])
            text = text.replace("【T】", f"{data_P:.2f}%")
            
        if "【AC】" in text:
            data_P = float(df[COL_NAMES["AC"]][i])
            text = text.replace("【AC】", f"{data_P:.2f}")
            
        if "【AB】" in text:
            data_P = float(df[COL_NAMES["AB"]][i])
            text = text.replace("【AB】", f"{data_P:.2f}")
            
        if "【TODAY】" in text:
            today = datetime.today()
            year = today.year
            month = today.month
            day = today.day
            text = text.replace("【TODAY】", f"{year}年{month}月{day}日")
            
        doc.paragraphs[j].text = ''
        
        # 加run用于修改字体
        run = doc.paragraphs[j].add_run(text)
        
        if j == 0:
            font_size = 18
            run.bold = True
        else:
            font_size = 14
            
        run.font.name = "Arial"
        run.font.size = Pt(font_size)
        run._element.rPr.rFonts.set(qn('w:eastAsia'),'宋体')
            
        # 调整字体
        # doc.paragraphs[j].style.font.size = Pt(font_size)

        # print(doc.paragraphs[j].text)
        
    doc.save(f"output_起诉状/{file_name}.docx")
```
